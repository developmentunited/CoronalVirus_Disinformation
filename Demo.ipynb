```python
import ktrain
from ktrain import text as txt
import numpy as np
```

```python
path = './all.tsv'
x_raw = []
y_raw = []
with open(path, 'rb') as inf:
    for line in inf:
        gzip_fields = line.decode('utf-8').split('\t')
        x_raw.append(gzip_fields[0])
        y_raw.append(gzip_fields[1].strip())
x_raw = np.array(x_raw)
y_raw = np.array(y_raw)
```

```python
train, test, preproc = txt.texts_from_array(x_train=x_raw,y_train=y_raw, 
                                                                   preprocess_mode='distilbert',maxlen=400,
                                                                   max_features=35000,val_pct=0.1)
```

```python
model = txt.text_classifier('distilbert', train_data=train, preproc=preproc)
```

```python
learner = ktrain.get_learner(model, train_data=train, val_data=test, batch_size=6)
learner.fit_onecycle(3e-5, 4)
```

```python
p = ktrain.get_predictor(model, preproc)

doc = 'A new alcohol-free sanitizer has been developed by the Dedan Kimathi University in Kenya.\
       The hand sanitizer developed by the Dedan Kimathi University actually contains alcohol.'

```

```python
p.explain(doc, all_targets=True)
```
